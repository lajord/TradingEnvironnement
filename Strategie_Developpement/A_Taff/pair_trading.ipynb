```python
# Import data

EURUSD = us.get_data_forex(['EURUSD'],['D1'],"2005-01-01","2010-01-01")
GPDUSD = us.get_data_forex(['GPDUSD'],['D1'],"2005-01-01","2010-01-01")
```

```python
import pandas as pd
import plotly.graph_objects as go
close_eurusd = EURUSD['D1']['EURUSD']['close'].to_frame().rename(columns={'close': 'EURUSD'})
close_gpdusd = GPDUSD['D1']['GPDUSD']['close'].to_frame().rename(columns={'close': 'GPDUSD'})
df = pd.concat([close_eurusd, close_gpdusd], axis=1).dropna()
df['Spread'] = df['EURUSD'] - df['GPDUSD']
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df.index,
    y=df['Spread'],
    mode='lines',
    name='Spread EURUSD/GPDUSD'
))
fig.update_layout(title='Spread EURUSD/GPDUSD', xaxis_title="Date",yaxis_title='Spread',template='plotly_dark')
fig.show()
```
