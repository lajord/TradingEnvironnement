```python
tickers2 = ["SP_TOTAL_4H","NK_TOTAL_4H","FDAX_TOTAL_4H","YM_TOTAL_4H"]
data2 = fe.get_data(tickers2, "2014-01-01", "2024-11-10")


strategie = tf.VpnFollow(data2,"240m",tickers2)

param1 = np.arange(2,30,1)
param2 = np.arange(5,30,1)
df = strategie.optimize(3,4,2,2,param1,param2,30,30,10,15)
```

```python
print([col[0] for col in df.columns] )
print([col[1] for col in df.columns])
```

```python

# Extraction des données
x1 = [col[0] for col in df.columns]  # Extraction des entiers pour X
z = [col[1] for col in df.columns]   # Extraction des entiers pour Z
y = df.iloc[0].values     

# Création des grilles pour ajouter des surfaces
x_unique = np.unique(x1)
z_unique = np.unique(z)
X, Z = np.meshgrid(x_unique, z_unique)

# Création de Y (valeurs interpolées pour la surface)
Y = np.full_like(X, np.nan, dtype=float)  # Initialisation avec des NaN
for i in range(len(x1)):
    xi = np.where(x_unique == x1[i])[0][0]
    zi = np.where(z_unique == z[i])[0][0]
    Y[zi, xi] = y[i]
Y = np.nan_to_num(Y, nan=0.0)  # Remplace les NaN par 0 pour éviter les erreurs

# Création du graphique 3D
fig = plt.figure(figsize=(15, 10))  # Taille élargie du graphique
ax = fig.add_subplot(111, projection='3d')

# Ajout de la surface continue
surf = ax.plot_surface(
    np.arange(len(x_unique)),
    np.arange(len(z_unique))[:, None],
    Y,
    cmap='viridis',
    alpha=0.8,
    edgecolor='k'
)

# Ajustement des ticks pour correspondre aux labels
ax.set_xticks(range(len(x_unique)))
ax.set_xticklabels(x_unique)
ax.set_yticks(range(len(z_unique)))
ax.set_yticklabels(z_unique)

# Ajout des étiquettes
ax.set_xlabel("X1 (Premier élément des tuples)", labelpad=20)
ax.set_ylabel("Z (Entiers)", labelpad=20)
ax.set_zlabel("Ratio de Calmar", labelpad=20)

# Configuration de l'angle de vue
ax.view_init(elev=90, azim=0)  # Exemple : élévation à 45° et azimut à 120°

# Ajout de marges pour les axes X et Z
ax.set_xlim(-0.5, len(x_unique) - 0.5)  # Ajuste les marges sur X
ax.set_ylim(-0.5, len(z_unique) - 0.5)  # Ajuste les marges sur Z

# Ajout de la barre de couleur
fig.colorbar(surf, ax=ax, shrink=0.8, aspect=5)

# Affichage du graphique
plt.title("Graphique 3D des Ratios de Calmar avec Surface Continue")
plt.show()
```
