```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Code pour remplacer les ; par des ,
df = pd.read_csv(r'C:\Users\Jordi\Desktop\Trading_Dev_Stratégie_Environement\Data\NQ\NQ-4H\nq-4h\nq-4h.csv', sep=';')
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

df['open'] = pd.to_numeric(df['open'], errors='coerce')
df['high'] = pd.to_numeric(df['high'], errors='coerce')
df['low'] = pd.to_numeric(df['low'], errors='coerce')
df['close'] = pd.to_numeric(df['close'], errors='coerce')
df['volume'] = pd.to_numeric(df['volume'], errors='coerce')

df.to_csv(f'SP_TOTAL.csv', index=False, sep=',')
```

```python
df = pd.read_csv(r'C:\Users\Jordi\Desktop\Trading_Dev_Stratégie_Environement\DossierUtile\FileTraitement\NQ4H_Total.csv', sep=',')
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Grouper les données par année
for annee, group in df.groupby(pd.Grouper(key='Date', freq='Y')):
    # Extraire l'année et sauvegarder dans un fichier CSV
    annee_str = annee.year
    group.to_csv(f'NQ4H_{annee_str}.csv', index=False, sep=',')
```
